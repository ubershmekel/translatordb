```python
import matplotlib.pyplot as plt

examples = list(tatoeba.generate_all_examples())

# Most sentences are 60 characters or shorter. Nice.
sentence_lens = [len(next(iter(it["eng"]))) for it in examples]
plt.hist(sentence_lens, bins=range(0, 150, 5))
plt.show()
```

```python
lang_counts = [len(it) for it in examples]
plt.hist(lang_counts, bins=range(0, 20, 1))
plt.show()
from collections import Counter
# Counter({2: 367379, 3: 226350, 4: 147584, 1: 104924, 5: 96734, 6: 65207, 7: 48457, 8: 31556, 9: 17222, 10: 7586, 11: 2878, 12: 884})
print(Counter(lang_counts))
```

```python
short_sentence_lang_counts = [len(it) for it in examples if len(next(iter(it["eng"]))) < 60]
plt.hist(short_sentence_lang_counts, bins=range(0, 20, 1))
plt.show()
from collections import Counter
# Counter({2: 367379, 3: 226350, 4: 147584, 1: 104924, 5: 96734, 6: 65207, 7: 48457, 8: 31556, 9: 17222, 10: 7586, 11: 2878, 12: 884})
print(Counter(short_sentence_lang_counts))
```
